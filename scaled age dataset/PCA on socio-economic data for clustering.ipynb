## Socio-economic dataset preprocessing 2

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_excel('Asansol socio-economic data 1.xlsx')
df = df.drop(columns=['ID','Unnamed: 20','Unnamed: 21','Unnamed: 22'])
df
```

```python
minmax_scaler = MinMaxScaler()
df['Age_minmax_scaled'] = minmax_scaler.fit_transform(df[['Age (Yrs)']])
df
```

```python
df_1 = df.drop(columns = ['Age (Yrs)'])
df_1

list1 = list(df_1.columns)
df_1 = df_1[[list1[-1]]+list1[0:-1]]
df_1

df_2 = df_1.to_numpy()
df_2 = df_2.T
df_2
```

```python
from sklearn.decomposition import PCA
```

```python
pca_1 = PCA(n_components=19, copy=True, whiten=False, svd_solver='full', tol=0.0, iterated_power='auto', n_oversamples=10, power_iteration_normalizer='auto', random_state=None)
pca_1.fit_transform(df_2)

pca_1.explained_variance_ratio_

variance = pd.DataFrame(pca_1.explained_variance_ratio_,columns = ['variance'])
variance
```

```python
import numpy as np
pca_1.components_.shape
```

```python
components = pca_1.components_[0:10,:]
components.shape
```

```python
components = components.T
components
```

```python
df_pc = pd.DataFrame(components,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])
df_pc
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scores = {}

for i in range(2,20):
    kmeans = KMeans(n_clusters=i,random_state=42,n_init=10)
    kmeans.fit(df_pc)
    score = silhouette_score(df_pc,kmeans.labels_)
    scores[i] = score
    
df_scores = pd.DataFrame(list(scores.items()), columns=['no. of clusters', 'silhouette score'])
df_scores
```
