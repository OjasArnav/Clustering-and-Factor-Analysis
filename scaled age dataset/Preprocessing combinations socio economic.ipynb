# Data Preprocessing of socio-economic dataset

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df = pd.read_excel('Asansol socio-economic data 1.xlsx')
df = df.drop(columns=['Unnamed: 20','Unnamed: 21','Unnamed: 22'])
df

new_df = df.drop(columns=['ID'])
new_df
```

```python
age = new_df[['Age (Yrs)']]
age

scaler = MinMaxScaler(feature_range=(0,1),clip=False)

scaled_age = scaler.fit_transform(age)

scaled_age

norm_age = pd.DataFrame(scaled_age,columns=['Normalised Age (Yrs)'])
norm_age
```

```python
norm_age.describe()
```

```python
new_df['Normalised Age (Yrs)'] = norm_age

new_df
```

```python
df_copy = new_df.drop(columns = ['Age (Yrs)'])
df_copy
```

```python
list1 = list(df_copy.columns)

df_copy = df_copy[[list1[-1]]+list1[0:18]]

df_copy
```

```python
df_copy.to_excel('Scaled Age socio economic data.xlsx')
```

## Reducing the no. of columns 

### Education Columns

```python
list2 = list(df_copy.columns)
print(list2)
```

```python
df_edu = df_copy[list2[2:6]]
df_edu
```

```python
df_edu_1 = pd.DataFrame()

df_edu_1['Education: Not Graduated'] = df_edu['Education (<Class 10)'] + df_edu['Education (Class 10-12)']

df_edu_1['Education: Graduated'] = df_edu['Graduate'] + df_edu['Higher Edu']

df_edu_1
```

### Income Columns

```python
df_income = df_copy[list2[6:11]]
df_income
```

```python
df_income_1 = pd.DataFrame()

df_income_1['Income 0-40k'] = df_income['Income <20k'] + df_income['Income 20-40k']
df_income_1['Income >=40k'] = df_income['Income 40-60k'] + df_income['Income 60-80k'] + df_income['Income >80k']

df_income_1
```

```python
df_income_2 = pd.DataFrame()

df_income_2['Income 0-60k'] = df_income['Income <20k'] + df_income['Income 20-40k'] + df_income['Income 40-60k']
df_income_2['Income >=60k'] = df_income['Income 60-80k'] + df_income['Income >80k']

df_income_2
```

```python
df_income_3 = pd.DataFrame()

df_income_3['Income <40k'] = df_income['Income <20k'] + df_income['Income 20-40k']
df_income_3['Income 40-60k'] = df_income['Income 40-60k']
df_income_3['Income >60k'] = df_income['Income 60-80k'] + df_income['Income >80k']

df_income_3
```

### Experience Columns

```python
df_experience = df_copy[list2[11:18]]

df_experience
```

```python
df_experience_1 = pd.DataFrame()

df_experience_1['Experience (<5yr)'] = df_experience['Experience (<1yr)'] + df_experience['Experience (1-2yr)'] + df_experience['Experience (2-5yr)']
df_experience_1['Experience (5-20yr)'] = df_experience['Experience (5-10yr)'] + df_experience['Experience (10-20yr)']
df_experience_1['Experience (>20yr)'] = df_experience['Experience (20-30yr)'] + df_experience['Experience (>30yr)']

df_experience_1
```

```python
df_experience_2 = pd.DataFrame()

df_experience_2['Experience (<10yr)'] = df_experience['Experience (<1yr)'] + df_experience['Experience (1-2yr)'] + df_experience['Experience (2-5yr)'] + df_experience['Experience (5-10yr)']
df_experience_2['Experience (>=10yr)'] = df_experience['Experience (10-20yr)'] + df_experience['Experience (20-30yr)'] + df_experience['Experience (>30yr)']

df_experience_2
```

## New datasets for clustering

```python
df_0 = df_copy.copy()

df_0 = df_0.drop(columns=list2[2:18])

df_1 = df_0.copy()

df_1
```

```python
df_1 = pd.concat([df_1,df_edu_1,df_income_1,df_experience_1],axis=1)
df_1
```

```python
scores = {}

for i in range(2,55):
  kmeans = KMeans(n_clusters=i,random_state=42,n_init=10) 
  kmeans.fit(df_1)
  score = silhouette_score(df_1,kmeans.labels_) 
  scores[i] = score
df_scores = pd.DataFrame(list(scores.items()), columns=['no. of clusters', 'silhouette score']) 
df_scores
```

```python
df_2 = df_0.copy()

df_2 = pd.concat([df_2,df_edu_1,df_income_2,df_experience_1],axis=1)

df_2
```

```python
scores = {}

for i in range(2,55):
  kmeans = KMeans(n_clusters=i,random_state=42,n_init=10) 
  kmeans.fit(df_2)
  score = silhouette_score(df_2,kmeans.labels_) 
  scores[i] = score
df_scores = pd.DataFrame(list(scores.items()), columns=['no. of clusters', 'silhouette score']) 
df_scores
```
